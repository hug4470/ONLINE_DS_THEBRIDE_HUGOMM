# AGRESIONES LGTBIFÓBICAS EN ESPAÑA. 
## Evolución y análisis de subgrupos.

### TEMA

Este proyecto tiene como objetivo llevar a cabo un análisis exhaustivo de los delitos de odio registrados en España desde 2014 hasta 2023, segmentado por comunidades autónomas. La meta es identificar tendencias, patrones y variaciones en la incidencia de estos delitos a lo largo del tiempo y en diferentes regiones del país. Además, se buscará enriquecer el análisis con un dataset que incluya diversas identidades y realidades del colectivo LGTBIQ+.

### HIPÓTESIS

HIPÓTESIS PRINCIPAL: El aumento de las agresiones al colectivo LGTBIQ+ en el estado español.

SUB-HIPÓTESIS:

- Incremento de Delitos de Odio: Se hipotetiza que ha habido un aumento significativo en la cantidad de delitos de odio en España entre 2014 y 2023.

- Variabilidad por Comunidades Autónomas: Se plantea la hipótesis de que existen diferencias notables en la incidencia de delitos de odio entre comunidades autónomas. Confirmar o desmentir los sesgos centralistas vs. sesgos provinciales.
- Identificación de Subgrupos Vulnerables: Se sugiere que dentro del colectivo LGTBIQ+, ciertos subgrupos (como personas trans, no binarias, bisexuales, asexuales) son más propensos a sufrir delitos de odio, lo que resalta la necesidad de un análisis detallado sobre las realidades y experiencias específicas de estos grupos.

- Impacto de la Legislación: Se plantea la hipótesis de que la implementación de políticas públicas y leyes contra la discriminación en ciertas comunidades autónomas está asociada con una disminución de los delitos de odio en esas áreas, sugiriendo que las medidas legislativas pueden ser efectivas en la reducción de la violencia. También el análisis de los años en los que ha habido elecciones estatales y autonómicas con fuertes impactos.

```python
import folium
import json
import pandas as pd

df_mapa_dict = df_23.set_index("comunidad")["2023"].to_dict()

mapa_espana = folium.Map(location=[40.4637, -3.7492], zoom_start=6)

with open('data/comunidades.geojson') as f:
    geojson_data = json.load(f)

def estilo(feature):
    comunidad_nombre = feature['properties']['acom_name']
    valor = df_mapa_dict.get(comunidad_nombre, 0) 
    print(f"Comunidad: {comunidad_nombre}, Valor: {valor}") 
    fill_color = (
        'green' if valor > 10 else
        'yellow' if valor > 50 else
        'pink' if valor > 100 else
        'turquoise' if valor > 150 else
        'red'
    )
    return {
        'fillColor': fill_color,
        'color': 'black',
        'weight': 1,
        'fillOpacity': 0.5,
    }

folium.GeoJson(
    geojson_data,
    style_function=estilo,
    acom_name='geojson'
).add_to(mapa_espana)

folium.LayerControl().add_to(mapa_espana)

mapa_espana
```

## CAMBIOS LEGISTLATIVOS Y AÑOS

# 2023
Junio<br>
Ley de Igualdad LGTBI (la famosa ley trans):

Objetivo: Garantizar la igualdad de derechos y la protección del colectivo LGTBI.
Medidas:
Prohibición de discriminación por orientación sexual o identidad de género.
Creación de un registro estatal de delitos de odio.
Impulso de campañas de sensibilización y formación sobre diversidad sexual.
Protección específica para menores LGTBI, incluyendo el derecho a ser atendidos en centros educativos con respeto a su identidad.

# 2022
Diciembre <br>
Modificación del Código Penal:

Cambios:
Aumento de las penas para los delitos de odio basados en la orientación sexual o identidad de género.
Introducción de medidas para mejorar la respuesta de las fuerzas de seguridad y del sistema judicial ante estos delitos.

# 2021
Junio<br>
1. Ley Orgánica de Protección de la Infancia y la Adolescencia frente a la Violencia:<br>
Contenido: Se incluye un enfoque específico sobre la protección de los menores LGTBI frente a la violencia y la discriminación, con medidas de prevención y atención.

Febrero<br>
2. Plan de Acción para la Igualdad LGTBI:<br>
Objetivo: Desarrollar políticas públicas para erradicar la discriminación y promover la igualdad, con un enfoque en la violencia y los crímenes de odio.
