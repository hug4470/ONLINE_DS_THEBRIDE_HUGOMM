```python
plt.figure(figsize=(10, 6))
sns.swarmplot(pokemonmelted, x = "Type 1", y = "value", dodge = True, palette= colores);
plt.ylim(0,200)
plt.legend("Dato", loc= 2, bbox_to_anchor=(0, 1));
```
