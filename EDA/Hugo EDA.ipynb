# AGRESIONES LGTBIFÓBICAS EN ESPAÑA. 
## Evolución y análisis de subgrupos.

### TEMA

Este proyecto tiene como objetivo llevar a cabo un análisis exhaustivo de los delitos de odio registrados en España desde 2014 hasta 2023, segmentado por comunidades autónomas. La meta es identificar tendencias, patrones y variaciones en la incidencia de estos delitos a lo largo del tiempo y en diferentes regiones del país. Además, se buscará enriquecer el análisis con un dataset que incluya diversas identidades y realidades del colectivo LGTBIQ+.

### HIPÓTESIS

HIPÓTESIS PRINCIPAL: El aumento de las agresiones al colectivo LGTBIQ+ en el estado español.

SUB-HIPÓTESIS:

- Incremento de Delitos de Odio: Se hipotetiza que ha habido un aumento significativo en la cantidad de delitos de odio en España entre 2014 y 2023.

- Variabilidad por Comunidades Autónomas: Se plantea la hipótesis de que existen diferencias notables en la incidencia de delitos de odio entre comunidades autónomas.
- Identificación de Subgrupos Vulnerables: Se sugiere que dentro del colectivo LGTBIQ+, ciertos subgrupos (como personas trans, no binarias, bisexuales, asexuales) son más propensos a experimentar delitos de odio, lo que resalta la necesidad de un análisis detallado sobre las realidades y experiencias específicas de estos grupos.

- Impacto de la Legislación: Se plantea la hipótesis de que la implementación de políticas públicas y leyes contra la discriminación en ciertas comunidades autónomas está asociada con una disminución de los delitos de odio en esas áreas, sugiriendo que las medidas legislativas pueden ser efectivas en la reducción de la violencia.
